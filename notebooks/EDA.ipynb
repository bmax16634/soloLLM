```python
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_from_disk

# Load your tokenized dataset
dataset = load_from_disk("../data/openwebtext_tokenized")

# Get lengths of each document
lengths = [len(ex["input_ids"]) for ex in dataset]

print(f"Total docs: {len(lengths)}")
print(f"Min length: {np.min(lengths)}")
print(f"Max length: {np.max(lengths)}")
print(f"Mean length: {np.mean(lengths):.1f}")
print(f"Median length: {np.median(lengths)}")

# Percentiles give a sense of distribution
for p in [50, 75, 90, 95, 99]:
    print(f"{p}th percentile length: {np.percentile(lengths, p):.0f}")

# Plot histogram (cut off at a reasonable length for clarity)
plt.hist([l for l in lengths if l < 2000], bins=100)
plt.xlabel("Document length (tokens)")
plt.ylabel("Count")
plt.title("Distribution of tokenized doc lengths")
plt.show()
```
